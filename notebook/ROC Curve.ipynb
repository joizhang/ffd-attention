```python
import os
import pickle
import time

import matplotlib.pyplot as plt
from sklearn.metrics import auc, roc_curve
```

```python
with open('../plot/xception.pickle', "rb") as f:
    y_true, y_pred, y_score = pickle.load(f)
print(len(y_true), len(y_pred), len(y_score))
fpr1, tpr1, thresholds1 = roc_curve(y_true, y_score, drop_intermediate=False)
roc_auc1 = auc(fpr1, tpr1)
```

```python
with open('../plot/vgg16.pickle', "rb") as f:
    y_true, y_pred, y_score = pickle.load(f)
print(len(y_true), len(y_pred), len(y_score))
fpr2, tpr2, thresholds2 = roc_curve(y_true, y_score, drop_intermediate=False)
roc_auc2 = auc(fpr2, tpr2)
```

```python
plt.plot(fpr1, tpr1, 'r-', label='Xception (AUC = {:.4f})'.format(roc_auc1))
plt.plot(fpr2, tpr2, 'b-', label='VGG16 (AUC = {:.4f})'.format(roc_auc2))
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.show()
```
